```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\amith\\Documents\\GitHub\\kaggle_2\\Quality of Life Index by Country\\quality_of_life_indices_by_country.csv")
df["Climate Index"] = df["Climate Index"].str.replace("-","0")
df["Climate Index"] = df["Climate Index"].astype("float64")
df.head()
```

```python
df_overall = df.groupby("Country")[["Quality of Life Index", "Purchasing Power Index", "Safety Index", "Health Care Index", "Cost of Living Index", "Property Price to Income Ratio", "Traffic Commute Time Index", "Pollution Index", "Climate Index"]].mean().round(2).reset_index()
df_overall = df_overall.sort_values("Quality of Life Index",ascending=False).reset_index(drop=True).reset_index()
df_overall = df_overall.rename(columns={"index":"Rank"})
df_overall["Rank"] = df_overall["Rank"] + 1
df_overall.head(10)
```
